```python
import urllib
import json                 # Used to load data into JSON format
from pprint import pprint   # pretty-print
```

```python
url = "https://aoe2.net/api/leaderboard?game=aoe2de&leaderboard_id=3&start=1&count=10000"
response = urllib.request.urlopen(url)
print(response)
# Just an object:   <http.client.HTTPResponse at 0x7f2618123e10>
```

```python
text = response.read()
```

```python
json_data = json.loads(text)
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(data=json_data['leaderboard'])
df1 = df1.set_index('rank')
df1.tail()
```
